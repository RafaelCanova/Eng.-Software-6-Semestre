# Import de Bibliotecas e Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

# Dados e Atributos

```python
df = load_iris()
print(df)
x = df.data[:, [2, 3]]
y = df.target
```

# Separação Treino/Teste

```python
x_test, x_train, y_test, y_train = train_test_split(x, y, test_size=0.2, random_state = 42)
```

# Treino do modelo

```python
svm = SVC(kernel='linear')
svm.fit(x_train, y_train)
```

# Resutados

```python
print(f'Acurácia: {accuracy_score(y_test, svm.predict(x_test))}')
```

# Comentários

O modelo apresenta alta acuracidade ao analisar os dados do dataset, porém parte disso se deve ao dataset iris já ser conhecido e não apresentar dados ruidosos,logo o modelo não é tão confiavel caso seja usado para dados "reais" onde podem haver erros e ruído nos dados coletados além de que o modelo feito não passou por uma otimização de parâmetros.
