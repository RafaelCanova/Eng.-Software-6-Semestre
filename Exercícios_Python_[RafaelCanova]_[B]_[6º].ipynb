Exercício 20 (Matriz Transposta)

```python
def transpor_matriz(matriz):
    linhas = len(matriz)
    colunas = len(matriz[0])

    transposta = [[0] * linhas for _ in range(colunas)]

    for i in range(linhas):
        for j in range(colunas):
            transposta[j][i] = matriz[i][j]

    return transposta

matriz_original = [
    [1, 2, 3],
    [4, 5, 6]
]

matriz_transposta = transpor_matriz(matriz_original)

for linha in matriz_transposta:
    print(linha)
```

Exercício 22 (Criptografia da Cifra de César)

```python
from typing_extensions import Text
texto = input("Digite o texto a ser criptografado: ")
texto = Text(texto)
deslocamento = 5
def cifra_de_cesar(texto, deslocamento):
    resultado = ""
    for char in texto:
        if char.isalpha():
            deslocamento_real = deslocamento % 26
            codigo = ord(char) + deslocamento_real
            if char.islower():
                if codigo > ord('z'):
                    codigo -= 26
            elif char.isupper():
                if codigo > ord('Z'):
                    codigo -= 26
            resultado += chr(codigo)
        else:
            resultado += char
    return resultado
texto_criptografado = cifra_de_cesar(texto, deslocamento)
print("Texto criptografado:", texto_criptografado)
```

Exercício 23 (Descriptografia da Cifra de César)

```python
def decifra_de_cesar(texto, deslocamento):
    resultado = ""
    for char in texto:
        if char.isalpha():
            deslocamento_real = deslocamento % 26
            codigo = ord(char) - deslocamento_real
            if char.islower():
                if codigo < ord('a'):
                    codigo += 26
            elif char.isupper():
                if codigo < ord('A'):
                    codigo += 26
            resultado += chr(codigo)
        else:
            resultado += char
    return resultado
print(decifra_de_cesar(texto_criptografado, deslocamento))
```

Exercício 25 (Classificação de Números)

```python
def classificar_numeros(lista):
    positivos = []
    negativos = []
    zeros = []

    for num in lista:
        if num > 0:
            positivos.append(num)
        elif num < 0:
            negativos.append(num)
        else:
            zeros.append(num)

    return {
        "Positivos": positivos,
        "Negativos": negativos,
        "Zeros": zeros
    }
numeros = [4, -3, 0, 7, -1, 0, 8, -6]
resultado = classificar_numeros(numeros)
print(resultado)
```

Exercício 31 (Ordenação por Bolha)

```python
def bubble_sort(list):
    n = len(list)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if list[j] < list[j + 1]:
                list[j], list[j + 1] = list[j + 1], list[j]
    return list

test_list = [3, 1, 4, 2]
print(bubble_sort(test_list))
```

Exercício 32 (Jogo da Velha)

```python
def exibir_tabuleiro(tabuleiro):
    for linha in tabuleiro:
        print(" | ".join(linha))
        print("-" * 9)

def verificar_vitoria(tabuleiro, jogador):
    for i in range(3):
        if all(tabuleiro[i][j] == jogador for j in range(3)) or all(tabuleiro[j][i] == jogador for j in range(3)):
            return True

    if all(tabuleiro[i][i] == jogador for i in range(3)) or all(tabuleiro[i][2 - i] == jogador for i in range(3)):
        return True

    return False

def jogo_da_velha():
    tabuleiro = [[" " for _ in range(3)] for _ in range(3)]
    jogador_atual = "X"

    for _ in range(9):
        exibir_tabuleiro(tabuleiro)

        while True:
            try:
                linha = int(input(f"Jogador {jogador_atual}, escolha a linha (0, 1, 2): "))
                coluna = int(input(f"Jogador {jogador_atual}, escolha a coluna (0, 1, 2): "))
                if tabuleiro[linha][coluna] == " ":
                    tabuleiro[linha][coluna] = jogador_atual
                    break
                else:
                    print("Essa posição já está ocupada. Tente novamente.")
            except (ValueError, IndexError):
                print("Entrada inválida. Digite um número entre 0 e 2.")

        if verificar_vitoria(tabuleiro, jogador_atual):
            exibir_tabuleiro(tabuleiro)
            print(f"Jogador {jogador_atual} venceu!")
            return

        jogador_atual = "O" if jogador_atual == "X" else "X"

    exibir_tabuleiro(tabuleiro)
    print("Empate!")

jogo_da_velha()
```

```python
def remover_duplicatas(lista):
    lista_sem_duplicatas = []
    for elemento in lista:
        if elemento not in lista_sem_duplicatas:
            lista_sem_duplicatas.append(elemento)
    return lista_sem_duplicatas


lista = [1, 2, 2, 3, 4, 4, 5, 6, 6, 7, 8, 8, 9, 10, 10]
resultado = remover_duplicatas(lista)
print(resultado)
```

Exercício 47 (Contar Números em Lista)

```python
def contar_ocorrencias(lista):
    contagem = {}
    for numero in lista:
        if numero in contagem:
            contagem[numero] += 1
        else:
            contagem[numero] = 1
    return contagem

lista_numeros = [1, 2, 3, 1, 2, 1, 4, 5, 4, 3, 3, 3]
resultado = contar_ocorrencias(lista_numeros)

for numero, quantidade in resultado.items():
    print(f"O número {numero} aparece {quantidade} vez(es).")
```

Exercício 48 (Ordenação por Inserção)

```python
def insertionSort(arr):
    n = len(arr)

    if n <= 1:
        return

    for i in range(1, n):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key

arr = [12, 11, 13, 5, 6]
insertionSort(arr)
print(arr)
```

Exercício 50 (Calculo de Frequência de Letras)

```python
def contar_frequencia_letras(texto):
    texto = texto.lower()
    contador = {}
    for letra in texto:
        if letra.isalpha():
            contador[letra] = contador.get(letra, 0) + 1
    return contador

texto_usuario = input("Digite uma string: ")
frequencia = contar_frequencia_letras(texto_usuario)

for letra, quantidade in sorted(frequencia.items()):
    print(f"{letra}: {quantidade}")
```
